```python
def rhombus(rows):
    for i in range(1, rows // 2 + 2):  
        for j in range((rows // 2) - i + 2):
            print(" ", end="")

        for j in range(2 * i - 1):
            print("*", end=" ")
        print()

    for i in range(rows // 2, 0, -1): 
        for j in range((rows // 2) - i + 2):
            print(" ", end="")
        
        for j in range(2 * i - 1):
            print("*", end=" ")
        print()

rows = int(input("Enter an odd number: "))
if rows % 2 == 1:
    rhombus(rows)
else:
    print("Please enter an odd number.")
```
